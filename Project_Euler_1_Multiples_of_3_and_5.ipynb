```python
import sys
def sumoftf(x,n):
    divisor = (n - 1) // x
    return (divisor * x * (divisor + 1)) // 2

a=0
b=0
c=0
t = int(input().strip())
if t<=(10**5) and t>=1:
    for a0 in range(t):
        n = int(input().strip())
        a=sumoftf(3,n)
        b=sumoftf(5,n)
        c=sumoftf(15,n)
        print(a+b-c)
```
